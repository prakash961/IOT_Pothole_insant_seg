```python
from ultralytics import YOLO
```

```python
model=YOLO("/content/drive/MyDrive/best.pt")
```

```python
model.train(data="/content/drive/MyDrive/Pothole Project 2.v1i.yolov8/data.yaml",epochs=1)
```

```python
model.val()
```

```python
results=model.predict("/content/drive/MyDrive/Pothole Project 2.v1i.yolov8/test/images",show=True)
```
